```python
class CabinSplitter(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        # Fill missing Cabin values with np.nan for imputation later
        cabin_filled = X['Cabin'].fillna(np.nan)

        # Split only non-null cabin values
        # Use a temporary DataFrame for splitting to handle NaNs gracefully
        split_data = cabin_filled.str.split('/', expand=True)

        X['Deck'] = split_data[0]
        # Convert CabinNum to numeric, coercing errors to NaN
        X['CabinNum'] = pd.to_numeric(split_data[1], errors='coerce')
        X['Side'] = split_data[2]
        return X
```

```python
class PassengerSplitter(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        # Fill missing values with np.nan for imputation later
        passenger_id_filled = X['PassengerId'].fillna(np.nan)

        # Split into two parts: Group and Number
        # Split only non-null PassengerId values
        split_data = passenger_id_filled.str.split('_', expand=True)
        X['Group'] = split_data[0]
        X['Number'] = split_data[1]
        return X
```

```python
class FoodCourtUsed(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X['foodcourt_used'] = (X['FoodCourt'] > 0).astype(int)
        return X
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class FamilyGroupFeatures(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()

        # Extract surname
        X['surname'] = X['Name'].apply(
            lambda x: x.split()[1] if isinstance(x, str) else 'Unknown'
        )

        # Extract group
        X['group'] = X['PassengerId'].apply(lambda x: x.split('_')[0])

        # Family group size
        X['family_group_size'] = X.groupby(['group', 'surname'])['PassengerId'].transform('count')

        # Optional binary version
        X['same_family_group'] = (X['family_group_size'] > 1).astype(int)
        return X
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

class FastModelBasedCategoricalImputer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.models = {}
        self.encoders = {}
        self.cols_to_impute = None

    def fit(self, X, y=None):
        X = X.copy()

        # Select categorical columns
        self.cols_to_impute = X.select_dtypes(include='object').columns

        for col in self.cols_to_impute:

            if X[col].isnull().sum() == 0:
                continue

            train = X[X[col].notnull()]
            if train.empty:
                continue

            X_train = train.drop(columns=[col])
            y_train = train[col]

            # Encode features
            encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

            X_train_enc = X_train.copy()
            cat_cols = X_train_enc.select_dtypes(include='object').columns

            if len(cat_cols) > 0:
                X_train_enc[cat_cols] = encoder.fit_transform(X_train_enc[cat_cols])

            # Train small model (FAST)
            model = RandomForestClassifier(
                n_estimators=20,
                max_depth=5,
                n_jobs=-1,
                random_state=42
            )
            model.fit(X_train_enc, y_train)

            self.models[col] = model
            self.encoders[col] = encoder

        return self

    def transform(self, X):
        X = X.copy()

        for col in self.cols_to_impute:

            if col not in self.models:
                continue

            missing_mask = X[col].isnull()
            if missing_mask.sum() == 0:
                continue

            X_test = X.loc[missing_mask].drop(columns=[col])

            encoder = self.encoders[col]

            X_test_enc = X_test.copy()
            cat_cols = X_test_enc.select_dtypes(include='object').columns

            if len(cat_cols) > 0:
                X_test_enc[cat_cols] = encoder.transform(X_test_enc[cat_cols])

            preds = self.models[col].predict(X_test_enc)

            X.loc[missing_mask, col] = preds

        return X
```

```python
import numpy as np
from sklearn.preprocessing import FunctionTransformer
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd # Import pandas for DataFrame conversion

# Custom transformer for log1p with get_feature_names_out
class Log1pFeatureTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        # Use validate=False to allow transforming columns with non-positive values if they exist
        self.transformer = FunctionTransformer(np.log1p, validate=False)

    def fit(self, X, y=None):
        self.transformer.fit(X, y)
        return self

    def transform(self, X):
        return self.transformer.transform(X)

    def get_feature_names_out(self, input_features=None):
        if input_features is None:
            raise ValueError("input_features must be provided to get_feature_names_out for Log1pFeatureTransformer.")
        return [f"log1p_{col}" for col in input_features]

# Custom transformer to ensure non-negative values and provide get_feature_names_out
class ClipNonNegativeTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return np.maximum(X, 0)

    def get_feature_names_out(self, input_features=None):
        # This transformer does not change feature names, so it returns the input names
        if input_features is None:
            raise ValueError("input_features must be provided to get_feature_names_out for ClipNonNegativeTransformer.")
        return input_features

# Custom transformer to convert a NumPy array back to a DataFrame
class ArrayToDataFrameTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        if not isinstance(X, pd.DataFrame):
            if X.shape[1] != len(self.columns):
                raise ValueError(f"Input array has {X.shape[1]} columns, but expected {len(self.columns)}.")
            return pd.DataFrame(X, columns=self.columns)
        return X # If it's already a DataFrame, just pass it through

    def get_feature_names_out(self, input_features=None):
        # This transformer just converts the type, so feature names remain the same
        if input_features is None:
            # If input_features are not explicitly passed, use the columns defined during initialization
            return self.columns
        # Otherwise, assume input_features are the names of the columns being passed through
        return input_features

# Define numerical columns for multivariate imputation
numerical_cols_for_imputation = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'CabinNum',]
# Define spending columns for log1p transformation
spending_cols = ['RoomService', 'FoodCourt', 'Spa', 'VRDeck', 'ShoppingMall']

preprocessor = ColumnTransformer(transformers=[
    ('numerical_pipeline', Pipeline([
        ('imputer', IterativeImputer(random_state=42)), # Multivariate imputation for all numerical features
        ('fallback_imputer', SimpleImputer(strategy='mean')), # Add fallback imputer for any remaining NaNs after IterativeImputer
        ('array_to_df', ArrayToDataFrameTransformer(columns=numerical_cols_for_imputation)), # Convert array to DataFrame
        ('feature_transform', ColumnTransformer(transformers=[
            ('impute_and_log_spending', Pipeline([
                ('imputer_before_log', SimpleImputer(strategy='constant', fill_value=0)), # Ensure no NaNs before log1p, treat missing spending as 0
                ('clip_non_negative', ClipNonNegativeTransformer()), # Ensure values are non-negative
                ('log1p', Log1pFeatureTransformer())
            ]), spending_cols),
            ('passthrough_age_cabin', 'passthrough', [col for col in numerical_cols_for_imputation if col not in spending_cols])
        ], remainder='drop')),
        ('scaler', StandardScaler())
    ]), numerical_cols_for_imputation),
    ('cat_impute_encode', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')), # Most frequent for nominal categoricals
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ]), ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Deck', 'Side']),
    ('bin', 'passthrough', ['foodcourt_used']),
    ('drop_cols', 'drop', ['Cabin','PassengerId','Name', 'Group', 'Number']) # Explicitly drop new features from PassengerSplitter
],remainder='drop'
)
```
